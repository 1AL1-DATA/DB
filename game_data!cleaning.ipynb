```python
import pandas as pd
import re

# words you want to ignore in names
STOPWORDS = {"edition", "ed", "remastered", "deluxe", "the", "a", "game", "version"}

def canonicalize(name):
    name = name.lower()
    name = re.sub(r"[^\w\s]", " ", name)    # remove punctuation
    name = re.sub(r"\s+", " ", name).strip() # collapse spaces
    tokens = [t for t in name.split() if t not in STOPWORDS]
    tokens = sorted(tokens)  # sort alphabetically to handle reordering
    return " ".join(tokens)
```

```python
vgchartz_2024["canonical_name"] = vgchartz_2024["Game"].apply(canonicalize)
Twitch_game_data["canonical_name"] = Twitch_game_data["Game"].astype(str).apply(canonicalize)
```

```python
from rapidfuzz import process, fuzz

def fuzzy_match(row, choices):
    match, score, idx = process.extractOne(
        row["canonical_name"],
        choices,
        scorer=fuzz.token_sort_ratio
    )
    return pd.Series([match, score])

choices = Twitch_game_data["canonical_name"].tolist()

vgchartz_2024[["best_match", "match_score"]] = vgchartz_2024.apply(
    fuzzy_match,
    axis=1,
    choices=choices
)
```

```python
from rapidfuzz import process, fuzz
import pandas as pd

def fuzzy_match_games(vgchartz_df, twitch_df, threshold=85, limit=1):
    """
    Fuzzy match game names between vgchartz and twitch datasets

    Parameters:
    -----------
    vgchartz_df: DataFrame with 'Game' column
    twitch_df: DataFrame with 'Game' column
    threshold: Minimum similarity score (0-100)
    limit: Number of best matches to return
    """
    vg_games = vgchartz_2024['Game'].astype(str).tolist()
    twitch_games = Twitch_game_data["Game"].astype(str).tolist()

    matches = []

    for vg_game in vg_games:
        # Find best match
        result = process.extractOne(
            vg_game,
            twitch_games,
            scorer=fuzz.token_sort_ratio,
            score_cutoff=threshold
        )

        if result:
            matched_game, score, index = result
            matches.append({
                'vgchartz_game': vg_game,
                'twitch_game': matched_game,
                'similarity_score': score,
                'twitch_index': index
            })
        else:
            matches.append({
                'vgchartz_game': vg_game,
                'twitch_game': None,
                'similarity_score': None,
                'twitch_index': None
            })

    return pd.DataFrame(matches)

# Usage
matches_df = fuzzy_match_games(vgchartz_2024, Twitch_game_data, threshold=85)
print(f"Found {matches_df['twitch_game'].notna().sum()} matches out of {len(matches_df)}")
```

```python
import re
from rapidfuzz import process, fuzz
import pandas as pd
import unicodedata

class GameNormalizer:
    """Normalizes game names for better fuzzy matching"""

    @staticmethod
    def normalize_latin_numbers(text: str) -> str:
        """Convert Latin/Roman numerals to Arabic numbers"""
        if not isinstance(text, str):
            return ""

        # Dictionary for Roman numeral conversion
        roman_to_arabic = {
            ' I ': ' 1 ', ' II ': ' 2 ', ' III ': ' 3 ', ' IV ': ' 4 ', ' V ': ' 5 ',
            ' VI ': ' 6 ', ' VII ': ' 7 ', ' VIII ': ' 8 ', ' IX ': ' 9 ', ' X ': ' 10 ',
            ' XI ': ' 11 ', ' XII ': ' 12 ', ' XIII ': ' 13 ', ' XIV ': ' 14 ', ' XV ': ' 15 ',
            ' XVI ': ' 16 ', ' XVII ': ' 17 ', ' XVIII ': ' 18 ', ' XIX ': ' 19 ', ' XX ': ' 20 '
        }

        # Also handle without spaces
        for roman, arabic in list(roman_to_arabic.items()):
            roman_nospace = roman.strip()
            arabic_nospace = arabic.strip()
            text = re.sub(fr'\b{roman_nospace}\b', arabic_nospace, text)

        # Handle ordinal numbers
        ordinal_patterns = {
            r'(\d+)(?:st|nd|rd|th)\b': r'\1',  # 1st -> 1, 2nd -> 2
            r'\b(first|second|third)\b': lambda m: {'first': '1', 'second': '2', 'third': '3'}[m.group(0).lower()]      }

        for pattern, replacement in ordinal_patterns.items():
            text = re.sub(pattern, replacement, text, flags=re.IGNORECASE)

        return text

    @staticmethod
    def normalize_text(text: str) -> str:
        """Normalize text for comparison"""
        if not isinstance(text, str):
            return ""

        # Convert to lowercase
        text = text.lower()

        # Remove diacritics (accents)
        text = unicodedata.normalize('NFKD', text)
        text = ''.join([c for c in text if not unicodedata.combining(c)])

        # Remove special characters but keep numbers and basic punctuation
        text = re.sub(r'[^\w\s\-&+]', '', text)

        # Normalize whitespace
        text = ' '.join(text.split())

        # Remove common prefixes/suffixes
        prefixes = ['the ', 'a ', 'an ', 'new ', 'super ', 'ultra ']
        suffixes = [' edition', ' version', ' definitive edition', ' game of the year',
                   ' goty', ' remastered', ' remake', ' hd', ' 4k']

        for prefix in prefixes:
            if text.startswith(prefix):
                text = text[len(prefix):]

        for suffix in suffixes:
            if text.endswith(suffix):
                text = text[:-len(suffix)]

        return text.strip()

    @staticmethod
    def extract_year(text: str) -> str:
        """Extract year from text"""
        year_match = re.search(r'\b(19\d{2}|20\d{2})\b', text)
        return year_match.group(0) if year_match else None

    @staticmethod
    def normalize_game_name(game_name: str) -> str:
        """Full normalization pipeline for game names"""
        if not isinstance(game_name, str):
            return ""

        # Step 1: Normalize Latin/Roman numbers
        normalized = GameNormalizer.normalize_latin_numbers(game_name)

        # Step 2: Normalize text
        normalized = GameNormalizer.normalize_text(normalized)

        return normalized


class EnhancedFuzzyScorer:
    """Enhanced scoring with string length and number matching considerations"""

    @staticmethod
    def weighted_composite_scorer(s1: str, s2: str) -> float:
        """
        Composite scorer that considers multiple factors

        Weights can be adjusted based on importance:
        - token_sort_ratio: Good for word order differences
        - partial_ratio: Good for partial matches
        - token_set_ratio: Good for extra/missing words
        - number_similarity: For number matching
        - length_similarity: For length consideration
        """

        # Base fuzzy scores
        token_sort_score = fuzz.token_sort_ratio(s1, s2)
        partial_score = fuzz.partial_ratio(s1, s2)
        token_set_score = fuzz.token_set_ratio(s1, s2)

        # Extract numbers from strings
        s1_numbers = re.findall(r'\d+', s1)
        s2_numbers = re.findall(r'\d+', s2)

        # Number matching score
        number_score = EnhancedFuzzyScorer.calculate_number_similarity(s1_numbers, s2_numbers)

        # Length similarity score
        length_score = EnhancedFuzzyScorer.calculate_length_similarity(s1, s2)

        # Year matching bonus
        year_bonus = EnhancedFuzzyScorer.calculate_year_match_bonus(s1, s2)

        # Weighted composite score
        weights = {
            'token_sort': 0.30,
            'partial': 0.25,
            'token_set': 0.20,
            'numbers': 0.15,
            'length': 0.05,
            'year_bonus': 0.05
        }

        composite_score = (
            token_sort_score * weights['token_sort'] +
            partial_score * weights['partial'] +
            token_set_score * weights['token_set'] +
            number_score * weights['numbers'] +
            length_score * weights['length'] +
            year_bonus * weights['year_bonus']
        )

        return min(100, composite_score)  # Cap at 100

    @staticmethod
    def calculate_number_similarity(numbers1: list, numbers2: list) -> float:
        """Calculate similarity based on number matching"""
        if not numbers1 and not numbers2:
            return 100  # Both have no numbers

        if not numbers1 or not numbers2:
            return 0  # One has numbers, other doesn't

        # Convert to sets for comparison
        set1 = set(numbers1)
        set2 = set(numbers2)

        # Jaccard similarity for numbers
        intersection = len(set1.intersection(set2))
        union = len(set1.union(set2))

        if union == 0:
            return 100

        jaccard_similarity = (intersection / union) * 100

        # Additional weight for exact sequence match
        if numbers1 == numbers2:
            jaccard_similarity = min(100, jaccard_similarity + 20)

        return jaccard_similarity

    @staticmethod
    def calculate_length_similarity(s1: str, s2: str) -> float:
        """Calculate similarity based on string length"""
        len1 = len(s1)
        len2 = len(s2)

        if len1 == 0 and len2 == 0:
            return 100

        if len1 == 0 or len2 == 0:
            return 0

        # Calculate length ratio (closer to 1 is better)
        length_ratio = min(len1, len2) / max(len1, len2)

        # Convert to percentage (0-100 scale)
        return length_ratio * 100

    @staticmethod
    def calculate_year_match_bonus(s1: str, s2: str) -> float:
        """Bonus score for matching years"""
        year1 = GameNormalizer.extract_year(s1)
        year2 = GameNormalizer.extract_year(s2)

        if year1 and year2:
            if year1 == year2:
                return 20  # Bonus for exact year match
            else:
                return -10  # Penalty for different years

        return 0  # No years or only one has year


def enhanced_fuzzy_match_games(
    vgchartz_df: pd.DataFrame,
    twitch_df: pd.DataFrame,
    threshold: float = 85,
    use_enhanced_scorer: bool = True,
    normalize_names: bool = True
) -> pd.DataFrame:
    """
    Enhanced fuzzy match game names with normalization and improved matching

    Parameters:
    -----------
    vgchartz_df: DataFrame with 'Game' column
    twitch_df: DataFrame with 'Game' column
    threshold: Minimum similarity score (0-100)
    use_enhanced_scorer: Whether to use enhanced composite scorer
    normalize_names: Whether to normalize game names before matching

    Returns:
    --------
    DataFrame with match results including original and normalized names
    """

    # Extract and normalize game names
    if normalize_names:
        vgchartz_df = vgchartz_2024.copy()
        twitch_df = Twitch_game_data.copy()

        vgchartz_df['normalized_game'] = vgchartz_df['Game'].apply(
            GameNormalizer.normalize_game_name
        )
        twitch_df['normalized_game'] = twitch_df['Game'].apply(
            GameNormalizer.normalize_game_name
        )

        vg_games = vgchartz_df['normalized_game'].tolist()
        twitch_games = twitch_df['normalized_game'].tolist()
    else:
        vg_games = vgchartz_df['Game'].astype(str).tolist()
        twitch_games = twitch_df['Game'].astype(str).tolist()

    # Store original names
    vg_original = vgchartz_df['Game'].astype(str).tolist()
    twitch_original = twitch_df['Game'].astype(str).tolist()

    matches = []

    # Choose scorer
    if use_enhanced_scorer:
        scorer = lambda s1, s2, **kwargs: EnhancedFuzzyScorer.weighted_composite_scorer(s1, s2)
    else:
        scorer = fuzz.token_sort_ratio

    for idx, vg_game in enumerate(vg_games):
        # Find best match
        result = process.extractOne(
            vg_game,
            twitch_games,
            scorer=scorer,
            score_cutoff=threshold
        )

        if result:
            matched_game, score, match_idx = result

            # Calculate additional metrics
            original_vg = vg_original[idx]
            original_twitch = twitch_original[match_idx]

            # Get individual scores for analysis
            base_scores = {
                'token_sort': fuzz.token_sort_ratio(vg_game, matched_game),
                'partial': fuzz.partial_ratio(vg_game, matched_game),
                'token_set': fuzz.token_set_ratio(vg_game, matched_game)
            }

            # Extract numbers for analysis
            vg_numbers = re.findall(r'\d+', vg_game)
            twitch_numbers = re.findall(r'\d+', matched_game)

            matches.append({
                'vgchartz_game': original_vg,
                'twitch_game': original_twitch,
                'similarity_score': score,
                'twitch_index': match_idx,
                'vg_normalized': vg_game,
                'twitch_normalized': matched_game,
                'token_sort_score': base_scores['token_sort'],
                'partial_score': base_scores['partial'],
                'token_set_score': base_scores['token_set'],
                'vg_numbers': vg_numbers,
                'twitch_numbers': twitch_numbers,
                'numbers_match': vg_numbers == twitch_numbers,
                'length_vg': len(vg_game),
                'length_twitch': len(matched_game),
                'length_ratio': min(len(vg_game), len(matched_game)) / max(len(vg_game), len(matched_game)) if max(len(vg_game), len(matched_game)) > 0 else 0
            })
        else:
            matches.append({
                'vgchartz_game': vg_original[idx],
                'twitch_game': None,
                'similarity_score': None,
                'twitch_index': None,
                'vg_normalized': vg_game,
                'twitch_normalized': None,
                'token_sort_score': None,
                'partial_score': None,
                'token_set_score': None,
                'vg_numbers': None,
                'twitch_numbers': None,
                'numbers_match': None,
                'length_vg': None,
                'length_twitch': None,
                'length_ratio': None
            })

    return pd.DataFrame(matches)


def analyze_match_quality(matches_df: pd.DataFrame) -> pd.DataFrame:
    """Analyze and categorize match quality"""

    def categorize_match(row):
        if pd.isna(row['similarity_score']):
            return 'no_match'

        score = row['similarity_score']
        numbers_match = row['numbers_match']
        length_ratio = row['length_ratio']

        # High quality: high score AND numbers match AND similar length
        if score >= 90 and numbers_match and length_ratio >= 0.8:
            return 'excellent'

        # Good quality: decent score and reasonable metrics
        elif score >= 85 and (numbers_match or length_ratio >= 0.7):
            return 'good'

        # Fair quality: meets threshold but has issues
        elif score >= 80:
            return 'fair'

        # Low quality: barely passes threshold
        elif score >= 75:
            return 'low'

        else:
            return 'poor'

    matches_df['match_quality'] = matches_df.apply(categorize_match, axis=1)

    # Calculate summary statistics
    quality_counts = matches_df['match_quality'].value_counts()

    print("Match Quality Distribution:")
    for quality, count in quality_counts.items():
        percentage = (count / len(matches_df)) * 100
        print(f"  {quality}: {count} games ({percentage:.1f}%)")

    # Show examples of each quality level
    print("\nExamples by quality level:")
    for quality in ['excellent', 'good', 'fair', 'low', 'poor', 'no_match']:
        if quality in matches_df['match_quality'].values:
            example = matches_df[matches_df['match_quality'] == quality].iloc[0]
            print(f"\n{quality.upper()}:")
            print(f"  VGChartz: {example['vgchartz_game']}")
            print(f"  Twitch: {example.get('twitch_game', 'No match')}")
            print(f"  Score: {example.get('similarity_score', 'N/A')}")

    return matches_df


# Example usage
if __name__ == "__main__":


    vgchartz_df = pd.DataFrame(vgchartz_data)
    twitch_df = pd.DataFrame(twitch_data)

    # Run enhanced fuzzy matching
    matches_df = enhanced_fuzzy_match_games(
        vgchartz_df,
        twitch_df,
        threshold=75,
        use_enhanced_scorer=True,
        normalize_names=True
    )

    print("Enhanced Fuzzy Matching Results:")
    print("=" * 60)
    print(matches_df[['vgchartz_game', 'twitch_game', 'similarity_score']].to_string(index=False))

    # Analyze match quality
    matches_df = analyze_match_quality(matches_df)

    # Show detailed analysis for high-quality matches
    print("\n" + "=" * 60)
    print("DETAILED ANALYSIS FOR HIGH-QUALITY MATCHES:")
    print("=" * 60)

    high_quality = matches_df[matches_df['match_quality'].isin(['excellent', 'good'])]

    for _, row in high_quality.iterrows():
        print(f"\n{row['vgchartz_game']} → {row['twitch_game']}")
        print(f"  Overall Score: {row['similarity_score']:.1f}")
        print(f"  Token Sort: {row['token_sort_score']:.1f}")
        print(f"  Partial: {row['partial_score']:.1f}")
        print(f"  Token Set: {row['token_set_score']:.1f}")
        print(f"  Numbers: VG={row['vg_numbers']}, Twitch={row['twitch_numbers']}")
        print(f"  Length Ratio: {row['length_ratio']:.2f}")
        print(f"  Quality: {row['match_quality']}")
```
